Use yfinance to Extract Stock Data

```python
import yfinance as yf

# Get Tesla stock data
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='max')

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the data
tesla_data.to_csv('tesla_data.csv')

# Display the first five rows
print(tesla_data.head())
```

 Use Webscraping to Extract Tesla Revenue Data

```python
import requests
import bs4
import pandas as pd

# Get Tesla revenue data from Wikipedia
url = 'https://en.wikipedia.org/wiki/Tesla,_Inc.'
response = requests.get(url)

# Parse the HTML response
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# Find the table containing the Tesla revenue data
table = soup.find('table', class_='wikitable sortable')

# Extract the Tesla revenue data
tesla_revenue = []
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    tesla_revenue.append([columns[0].text, columns[1].text])

# Save the data as a pandas dataframe
tesla_revenue_df = pd.DataFrame(tesla_revenue, columns=['Date', 'Revenue'])
tesla_revenue_df.tail(5)
```

Plot Tesla Stock Graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Tesla stock data
tesla_data = pd.read_csv('tesla_data.csv')

# Make a graph of the Tesla stock price
plt.plot(tesla_data['Close'])

# Set the title and labels
plt.title('Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Show the graph
plt.show()
```
